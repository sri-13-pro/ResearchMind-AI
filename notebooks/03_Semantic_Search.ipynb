```python
import pandas as pd
import numpy as np

from sentence_transformers import SentenceTransformer
import faiss

print("Libraries imported successfully!")
```

```python
papers = pd.read_csv("../data/cleaned_papers.csv")

print("Dataset Loaded Successfully!")
papers.head()
```

```python
papers = papers[["title", "abstract"]]

papers.head()
```

```python
papers = papers.sample(5000, random_state=42).reset_index(drop=True)

print("Number of Papers:", len(papers))
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")

print("Model Loaded Successfully!")
```

```python
embeddings = model.encode(
    papers["abstract"].tolist(),
    show_progress_bar=True
)

print("Embeddings Shape:", embeddings.shape)
```

```python
embeddings = np.array(embeddings).astype("float32")
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(embeddings)

print("FAISS Index Created!")
print("Total Vectors:", index.ntotal)
```

```python
faiss.write_index(index, "../vector_db/research_index.faiss")

print("FAISS Index Saved!")
```

```python
def search_papers(query, top_k=5):
    # Convert query to embedding
    query_embedding = model.encode([query])

    query_embedding = np.array(query_embedding).astype("float32")

    distances, indices = index.search(query_embedding, top_k)

    results = papers.iloc[indices[0]][["title", "abstract"]]

    return results
```

```python
results = search_papers("Transformer models for NLP")

results
```

```python
results = search_papers("Deep Learning for Image Classification")

results
```

```python
results = search_papers("Computer Vision")

results
```

```python
results = search_papers("Large Language Models")

results
```
