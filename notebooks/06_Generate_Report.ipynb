- Paper Summary

- Reading Time

```python
import os
import sys

project_root = os.path.abspath("..")

if project_root not in sys.path:
    sys.path.insert(0, project_root)

from utils.pdf_utils import extract_text_from_pdf
from utils.summarizer import PaperSummarizer
from utils.analytics import PaperAnalytics
from utils.citation import CitationGenerator

from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer
)

from reportlab.lib.styles import getSampleStyleSheet

print("Modules Loaded Successfully!")
```

```python
pdf_path = "../uploads/attention_is_all_you_need.pdf"

paper = extract_text_from_pdf(pdf_path)

print("Paper Loaded!")
```

```python
summarizer = PaperSummarizer()

summary = summarizer.summarize(
    paper,
    sentences=8
)

keywords = summarizer.keywords(
    paper,
    top_n=10
)
```

```python
analytics = PaperAnalytics(paper)

stats = analytics.statistics()

freq = analytics.frequent_words(15)
```

```python
citation = CitationGenerator(
    title="Attention Is All You Need",
    authors="Ashish Vaswani et al.",
    year="2017"
)
```

```python
styles = getSampleStyleSheet()

doc = SimpleDocTemplate("../reports/Research_Report.pdf")

story = []
```

```python
story.append(
    Paragraph(
        "<b>ResearchMind AI Report</b>",
        styles["Title"]
    )
)

story.append(Spacer(1,20))
```

```python
story.append(
    Paragraph(
        "<b>Summary</b>",
        styles["Heading2"]
    )
)

for sentence in summary:
    story.append(
        Paragraph(sentence, styles["BodyText"])
    )
```

```python
story.append(Spacer(1,20))

story.append(
    Paragraph(
        "<b>Keywords</b>",
        styles["Heading2"]
    )
)

for keyword, score in keywords:
    story.append(
        Paragraph(
            f"• {keyword}",
            styles["BodyText"]
        )
    )
```

```python
story.append(Spacer(1,20))

story.append(
    Paragraph(
        "<b>Statistics</b>",
        styles["Heading2"]
    )
)

for key,value in stats.items():

    story.append(
        Paragraph(
            f"{key}: {value}",
            styles["BodyText"]
        )
    )
```

```python
story.append(Spacer(1,20))

story.append(
    Paragraph(
        "<b>Top Frequent Words</b>",
        styles["Heading2"]
    )
)

for _,row in freq.iterrows():

    story.append(
        Paragraph(
            f"{row['Word']} : {row['Frequency']}",
            styles["BodyText"]
        )
    )
```

```python
story.append(Spacer(1,20))

story.append(
    Paragraph(
        "<b>Citation (APA)</b>",
        styles["Heading2"]
    )
)

story.append(
    Paragraph(
        citation.apa(),
        styles["BodyText"]
    )
)
```

```python
doc.build(story)
```

```python
doc.build(story)

print("Report Generated Successfully!")
```
